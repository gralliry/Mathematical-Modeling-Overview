# 多元非线性回归

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# 定义非线性模型，例如多元指数模型
def exponential_model(X, beta0, beta1, beta2):
    X1, X2 = X
    return beta0 * np.exp(beta1 * X1) * np.exp(beta2 * X2)

# 示例数据
X_data = np.array([
    [1, 2],
    [2, 1],
    [3, 3],
    [4, 4],
    [5, 5]
])
y_data = np.array([2.3, 4.5, 9.1, 20.2, 42.3])

# 使用 curve_fit 进行多元非线性拟合
params, params_covariance = curve_fit(exponential_model, X_data.T, y_data, p0=[1, 0.5, 0.5])

# 提取拟合参数
beta0, beta1, beta2 = params

# 打印拟合参数
print(f'Fitted parameters: beta0 = {beta0}, beta1 = {beta1}, beta2 = {beta2}')

# 生成预测数据
x1_fit = np.linspace(min(X_data[:, 0]), max(X_data[:, 0]), 100)
x2_fit = np.linspace(min(X_data[:, 1]), max(X_data[:, 1]), 100)
X1_fit, X2_fit = np.meshgrid(x1_fit, x2_fit)
y_fit = exponential_model([X1_fit.ravel(), X2_fit.ravel()], *params).reshape(X1_fit.shape)

# 绘制结果
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_data[:, 0], X_data[:, 1], y_data, color='r', label='Data')
ax.plot_surface(X1_fit, X2_fit, y_fit, color='b', alpha=0.5, label='Fitted function')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
ax.legend()
plt.show()
```

# 决策树

```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np
import matplotlib.pyplot as plt

# 示例数据
X_data = np.array([[1, 2], [2, 1], [3, 3], [4, 4], [5, 5]])
y_data = np.array([2.3, 4.5, 9.1, 20.2, 42.3])

# 创建并训练决策树回归模型
model = DecisionTreeRegressor()
model.fit(X_data, y_data)

# 生成预测数据
x1_fit = np.linspace(min(X_data[:, 0]), max(X_data[:, 0]), 100)
x2_fit = np.linspace(min(X_data[:, 1]), max(X_data[:, 1]), 100)
X1_fit, X2_fit = np.meshgrid(x1_fit, x2_fit)
X_fit = np.vstack([X1_fit.ravel(), X2_fit.ravel()]).T
y_fit = model.predict(X_fit).reshape(X1_fit.shape)

# 绘制结果
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_data[:, 0], X_data[:, 1], y_data, color='r', label='Data')
ax.plot_surface(X1_fit, X2_fit, y_fit, color='b', alpha=0.5, label='Decision Tree fit')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
ax.legend()
plt.show()
```

# 向量机

```python
from sklearn.svm import SVR
import numpy as np
import matplotlib.pyplot as plt

# 示例数据
X_data = np.array([[1, 2], [2, 1], [3, 3], [4, 4], [5, 5]])
y_data = np.array([2.3, 4.5, 9.1, 20.2, 42.3])

# 创建并训练SVR模型
model = SVR(kernel='rbf', C=100, epsilon=0.1)
model.fit(X_data, y_data)

# 生成预测数据
x1_fit = np.linspace(min(X_data[:, 0]), max(X_data[:, 0]), 100)
x2_fit = np.linspace(min(X_data[:, 1]), max(X_data[:, 1]), 100)
X1_fit, X2_fit = np.meshgrid(x1_fit, x2_fit)
X_fit = np.vstack([X1_fit.ravel(), X2_fit.ravel()]).T
y_fit = model.predict(X_fit).reshape(X1_fit.shape)

# 绘制结果
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_data[:, 0], X_data[:, 1], y_data, color='r', label='Data')
ax.plot_surface(X1_fit, X2_fit, y_fit, color='b', alpha=0.5, label='SVR fit')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
ax.legend()
plt.show()
```

# 神经网络
