# 一元线性回归

```python
import numpy as np
import matplotlib.pyplot as plt

# 示例数据：身高（cm）和体重（kg）
heights = np.array([150, 160, 170, 180, 190])
weights = np.array([50, 60, 65, 70, 80])

# 添加一个常数列以计算截距项
X = np.vstack([heights, np.ones(len(heights))]).T

# 计算回归系数
# np.linalg.lstsq() 求解线性方程组的最小二乘解
coefficients, _, _, _ = np.linalg.lstsq(X, weights, rcond=None)

# 提取斜率和截距
slope, intercept = coefficients

# 打印回归系数
print(f'Slope: {slope}, Intercept: {intercept}')

# 预测体重
predicted_weights = slope * heights + intercept

# 绘制数据点和回归线
plt.scatter(heights, weights, color='blue', label='Data Points')
plt.plot(heights, predicted_weights, color='red', label='Regression Line')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# 示例数据：身高（cm）和体重（kg）
heights = np.array([150, 160, 170, 180, 190]).reshape(-1, 1)
weights = np.array([50, 60, 65, 70, 80])

# 创建线性回归模型
model = LinearRegression()

# 训练模型
model.fit(heights, weights)

# 获取回归系数和截距
slope = model.coef_[0]
intercept = model.intercept_

# 打印回归系数
print(f'Slope: {slope}, Intercept: {intercept}')

# 预测体重
predicted_weights = model.predict(heights)

# 判定系数
# 计算均值
mean_weight = np.mean(weights)

# 计算总平方和 (SST)
sst = np.sum((weights - mean_weight) ** 2)

# 计算回归平方和 (SSR)
sse = np.sum((predicted_weights - mean_weight) ** 2)

# 计算残差平方和 (SSE)
ssr = np.sum((weights - predicted_weights) ** 2)

# r2 = sse / sst
r2 = 1 - ssr / sst
# 0代表无关，1代表完全相关
print(r2)

# 显著性检验：1. 线性关系检验 2. slope回归系数检验

# 检验统计量
n = len(weights)
F = sse / (ssr / (n - 2))
print(F)
# F分布查表F(1, n-2)
# 如果大于则可接受
# 绘制数据点和回归线
plt.scatter(heights, weights, color='blue', label='Data Points')
plt.plot(heights, predicted_weights, color='red', label='Regression Line')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.legend()
plt.show()
```
