# 最小生成树 MST

n个点有n-1条边

# Kruskal 算法

```python
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib import rcParams

# 设置字体为 SimHei（支持中文）
rcParams['font.family'] = 'SimHei'

def kruskal_algorithm(graph):
    # 使用 NetworkX 的最小生成树算法
    T = nx.minimum_spanning_tree(graph)
    return T

# 创建图
G = nx.Graph()
edges = [(0, 1, 2), (1, 2, 3), (2, 0, 4), (1, 3, 5), (3, 4, 6), (4, 2, 7)]
G.add_weighted_edges_from(edges)

# 计算最小生成树
T = kruskal_algorithm(G)

# 绘制图形
pos = nx.spring_layout(G)
plt.figure(figsize=(12, 6))

# 绘制原图
plt.subplot(121)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700, font_size=16, font_weight='bold', edge_color='gray')
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title("原图")

# 绘制最小生成树
plt.subplot(122)
nx.draw(T, pos, with_labels=True, node_color='lightgreen', node_size=700, font_size=16, font_weight='bold', edge_color='black')
edge_labels_T = nx.get_edge_attributes(T, 'weight')
nx.draw_networkx_edge_labels(T, pos, edge_labels=edge_labels_T)
plt.title("最小生成树")

plt.show()
```
