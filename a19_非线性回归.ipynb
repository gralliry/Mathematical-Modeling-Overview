# 已知模型回归

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# 定义非线性模型，例如指数模型
def exponential_model(x, beta0, beta1):
    return beta0 * np.exp(beta1 * x)

# 示例数据
x_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([2.3, 4.5, 9.1, 20.2, 42.3])

# 使用curve_fit进行非线性拟合
params, params_covariance = curve_fit(exponential_model, x_data, y_data, p0=[1, 0.5])

# 提取拟合参数
beta0, beta1 = params

# 打印拟合参数
print(f'Fitted parameters: beta0 = {beta0}, beta1 = {beta1}')

# 绘制结果
plt.scatter(x_data, y_data, label='Data')
plt.plot(x_data, exponential_model(x_data, *params), color='red', label='Fitted function')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

## 多项式预测

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import Polynomial

# 示例数据
x_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([2.3, 4.5, 9.1, 20.2, 42.3])

# 使用多项式回归
degree = 3  # 多项式的度数
coefs = Polynomial.fit(x_data, y_data, degree).convert().coef

# 计算拟合值
x_fit = np.linspace(min(x_data), max(x_data), 100)
y_fit = np.polyval(coefs[::-1], x_fit)

# 绘制结果
plt.scatter(x_data, y_data, label='Data')
plt.plot(x_fit, y_fit, color='red', label='Polynomial fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

print(f'Fitted polynomial coefficients: {coefs}')
```

## 机器学习

```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np
import matplotlib.pyplot as plt

# 示例数据
x_data = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y_data = np.array([2.3, 4.5, 9.1, 20.2, 42.3])

# 创建并训练模型
model = DecisionTreeRegressor()
model.fit(x_data, y_data)

# 生成预测数据
x_fit = np.linspace(min(x_data), max(x_data), 100).reshape(-1, 1)
y_fit = model.predict(x_fit)

# 绘制结果
plt.scatter(x_data, y_data, label='Data')
plt.plot(x_fit, y_fit, color='red', label='Decision Tree fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

## 神经网络

```python
from sklearn.neural_network import MLPRegressor
import numpy as np
import matplotlib.pyplot as plt

# 示例数据
x_data = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y_data = np.array([2.3, 4.5, 9.1, 20.2, 42.3])

# 创建并训练神经网络模型
model = MLPRegressor(hidden_layer_sizes=(10,), max_iter=100000)
model.fit(x_data, y_data)

# 生成预测数据
x_fit = np.linspace(min(x_data), max(x_data), 100).reshape(-1, 1)
y_fit = model.predict(x_fit)

# 绘制结果
plt.scatter(x_data, y_data, label='Data')
plt.plot(x_fit, y_fit, color='red', label='Neural Network fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
