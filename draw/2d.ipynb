# 折线图

```python
import matplotlib.pyplot as plt

# 创建数据
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# 绘制折线图
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Line')

# 添加标题和标签
plt.title('Line Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()

plt.show()
```

# 条形图

```python
import matplotlib.pyplot as plt

# 创建数据
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 10]

# 绘制条形图
plt.bar(categories, values, color='g')

# 添加标题和标签
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.show()
```

```python
# 并排显示多个直方图
import matplotlib.pyplot as plt
import numpy as np

# 创建数据
labels = ['A', 'B', 'C', 'D']
values1 = [10, 15, 7, 10]
values2 = [5, 8, 12, 6]
values3 = [7, 10, 5, 9]

x = np.arange(len(labels))  # 标签位置
width = 0.25  # 条形的宽度

# 绘制直方图
fig, ax = plt.subplots()
bars1 = ax.bar(x - width, values1, width, label='Indicator 1')
bars2 = ax.bar(x, values2, width, label='Indicator 2')
bars3 = ax.bar(x + width, values3, width, label='Indicator 3')

# 添加标签和标题
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Multiple Indicators in Histogram')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```

```python
# 堆叠显示多个直方图
import matplotlib.pyplot as plt
import numpy as np

# 创建数据
labels = ['A', 'B', 'C', 'D']
values1 = [10, 15, 7, 10]
values2 = [5, 8, 12, 6]
values3 = [7, 10, 5, 9]

x = np.arange(len(labels))  # 标签位置
width = 0.5  # 条形的宽度

# 绘制直方图
fig, ax = plt.subplots()
bars1 = ax.bar(x, values1, width, label='Indicator 1')
bars2 = ax.bar(x, values2, width, bottom=values1, label='Indicator 2')
bars3 = ax.bar(x, values3, width, bottom=np.array(values1) + np.array(values2), label='Indicator 3')

# 添加标签和标题
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Stacked Histogram')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```

# 直方图

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建数据
data = np.random.randn(1000)

# 绘制直方图
plt.hist(data, bins=30, color='c', edgecolor='black')

# 添加标题和标签
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```

```python
# 多条直方图在同一图上
import matplotlib.pyplot as plt
import numpy as np

# 创建数据
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 2
data3 = np.random.randn(1000) - 2

# 绘制直方图
plt.hist(data1, bins=30, alpha=0.5, label='Data Set 1')
plt.hist(data2, bins=30, alpha=0.5, label='Data Set 2')
plt.hist(data3, bins=30, alpha=0.5, label='Data Set 3')

# 添加标签和标题
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Multiple Histograms on the Same Plot')
plt.legend()

plt.show()
```

```python
# 在子图中显示多个直方图
import matplotlib.pyplot as plt
import numpy as np

# 创建数据
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 2
data3 = np.random.randn(1000) - 2

fig, axs = plt.subplots(3, 1, figsize=(8, 12))

axs[0].hist(data1, bins=30, color='blue', alpha=0.7)
axs[0].set_title('Histogram for Data Set 1')

axs[1].hist(data2, bins=30, color='green', alpha=0.7)
axs[1].set_title('Histogram for Data Set 2')

axs[2].hist(data3, bins=30, color='red', alpha=0.7)
axs[2].set_title('Histogram for Data Set 3')

# 添加标签
for ax in axs:
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

# 饼图

```python
import matplotlib.pyplot as plt

# 创建数据
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']

# 绘制饼图
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# 添加标题
plt.title('Pie Chart')

plt.show()
```

# 等高线图

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建数据
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# 绘制等高线图
plt.contour(X, Y, Z, cmap='viridis')

# 添加标题和标签
plt.title('Contour Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

# 热图

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建数据
data = np.random.rand(10, 10)

# 绘制热图
plt.imshow(data, cmap='hot', interpolation='bilinear')
plt.colorbar()

# 添加标题
plt.title('Heatmap')

plt.show()
```

# 步骤图

```python
import matplotlib.pyplot as plt

# 创建数据
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# 绘制步骤图
plt.step(x, y, where='mid', color='purple', label='Step')

# 添加标题和标签
plt.title('Step Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()

plt.show()
```

# 面积图

```python
import matplotlib.pyplot as plt

# 创建数据
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# 绘制面积图
plt.fill_between(x, y, color='skyblue', alpha=0.4)

# 添加标题和标签
plt.title('Area Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```
