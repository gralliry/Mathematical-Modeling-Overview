# 有向图

```python
import networkx as nx
import matplotlib.pyplot as plt

# 创建有向图
G = nx.DiGraph()

# 添加节点
G.add_nodes_from([1, 2, 3, 4, 5])

# 添加有向边
G.add_edges_from([(1, 2), (1, 3), (2, 4), (3, 4), (4, 5)])

# 绘制有向图
pos = nx.spring_layout(G)  # 节点布局
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightblue', font_size=15, font_weight='bold', arrows=True)

# 显示图形
plt.show()
```

# 无向图

```python
import networkx as nx
import matplotlib.pyplot as plt

# 创建无向图
G = nx.Graph()

# 添加节点
G.add_nodes_from([1, 2, 3, 4, 5])

# 添加无向边
G.add_edges_from([(1, 2), (1, 3), (2, 4), (3, 4), (4, 5)])

# 绘制无向图
pos = nx.spring_layout(G)  # 节点布局
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightblue', font_size=15, font_weight='bold')

# 显示图形
plt.show()
```

# 混合向图

```python
import networkx as nx
import matplotlib.pyplot as plt

from matplotlib import rcParams

# 设置字体为 SimHei（支持中文）
rcParams['font.family'] = 'SimHei'

# 创建图
G = nx.DiGraph()

# 添加有向边
G.add_edge(0, 1, weight=2, style='dashed')
G.add_edge(1, 2, weight=3, style='dashed')
G.add_edge(2, 0, weight=4, style='dashed')

# 添加无向边
G.add_edge(0, 2, weight=1, style='solid')
G.add_edge(1, 3, weight=5, style='solid')

# 获取所有边的样式
edge_styles = nx.get_edge_attributes(G, 'style')

# 设置布局
pos = nx.spring_layout(G, seed=42)  # 使用 spring 布局

# 绘制所有节点
plt.figure(figsize=(12, 8))
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=700)

# 绘制有向边
directed_edges = [(u, v) for u, v, d in G.edges(data=True) if d.get('style') == 'dashed']
nx.draw_networkx_edges(G, pos, edgelist=directed_edges, arrowstyle='-|>', arrowsize=20, style='dashed', edge_color='gray')

# 绘制无向边
undirected_edges = [(u, v) for u, v, d in G.edges(data=True) if d.get('style') == 'solid']
nx.draw_networkx_edges(G, pos, edgelist=undirected_edges, style='solid', edge_color='black')

# 绘制所有边的标签
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# 绘制所有节点标签
nx.draw_networkx_labels(G, pos, font_size=16, font_weight='bold')

plt.title("有向边和无向边的图")
plt.show()
```
